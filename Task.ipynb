```python
import csv
```

```python
class ProcessingFileCSV:

    def __init__(self, filename):
        self.filename = filename


    # Чтение файла csv
    def read_csv(self):

        with open(self.filename, encoding='utf-8') as file:
            file_reader = csv.reader(file, delimiter = ",")

            count = 0
            abonents_list = list()
            for row in file_reader:
                if count == 0:
                    abonents_list.append((row[0]).split(';'))
                else:
                    abonent = row[0].split(';')
                    abonents_list.append(abonent)

                count += 1

        return abonents_list


    # Запись в файл csv
    def write_file(self, data_list, to_file_name):

        with open(to_file_name, mode='w', newline="") as w_file:
            file_writer = csv.writer(w_file, delimiter = ",")

            for row in data_list:
                file_writer.writerow(row)


    # Удаление колонок, аргумент cols должен быть
    # списком номеров колонок (начиная с 1)!
    def delete_cols(self, abonents_list, cols):

        for abonent in abonents_list:
            count = 0
            for col in cols:
                del abonent[col-count-1]
                count += 1
        
        return abonents_list




# Вычисление значения начисления для каждого абонента
def eval_accural_abonents(abonents_list):
    
    StNormative = '301.26'
    StRegister = 1.52

    for abonent in abonents_list:
        if abonent[5] == '1':
            abonent.append(StNormative)
        elif abonent[5] == '2':
            abonent.append(str((int(abonent[7]) - int(abonent[6]))*StRegister))

        # для строки с заголовками
        else:
            abonent[0] = abonent[0][1:]
            abonent.append('Начислено')
    
    return abonents_list


# Вычисление значения начисления для каждого дома
def eval_accural_houses(abonents_list):

    houses_dict = dict()
    for abonent in abonents_list[1:]:
        house = abonent[1] + ' ' + abonent[2]

        if house in houses_dict.keys():
            houses_dict[house] += float(abonent[3])
        else:
            houses_dict[house] = float(abonent[3])

    houses_list = [abonents_list[0]]
    count = 1
    for k,v in houses_dict.items():
        houses_list.append([count, *k.split(' '), v])
        count += 1
    
    return houses_list




# Функция для обработки файла согласно заданию 1
def processing1(filename):
    file = ProcessingFileCSV(filename)
    abonents_list = eval_accural_abonents(file.read_csv())
    file.write_file(abonents_list, 'Начисления_абоненты.csv')


# Функция для обработки файла согласно заданию 2
deleted_cols = [2,5,6,7,8]

def processing2(filename):
    file = ProcessingFileCSV(filename)
    abonents_list = file.delete_cols(eval_accural_abonents(file.read_csv()), deleted_cols)
    accural_houses = eval_accural_houses(abonents_list)
    file.write_file(accural_houses, "Начисления_дома.csv")



# ------------------------- #
#  Начало работы программы  #
# ------------------------- #

processing1('абоненты.csv')

processing2('абоненты.csv')
```
